# Determine minimal fuel needed to align crabs

```python
crabs_pos = list(map(int, open("input.txt", "r").read().split("\n")[0].split(",")))
```

```python
def c(x):
    return sum([abs(x - pi) for pi in crabs_pos])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(list(range(len(crabs_pos))), [c(i) for i in range(len(crabs_pos))])
```

```python
x = len(crabs_pos) // 2
while True:
    if c(x) > c(x - 1):
        x -= 1
    elif c(x) > c(x + 1):
        x += 1
    else:
        break

c(x)
```

# Challenge 2: irregular cost of the fuel

```python
lut = [0]
for i in range(1, 10000):
    lut.append(lut[-1] + i)
```

```python
lut[5]
```

```python
def c(x):
    s = 0
    for pi in crabs_pos:
        steps = abs(x - pi)
        cost = lut[steps]
        s += cost
    return s
```

```python
plt.plot(list(range(len(crabs_pos))), [c(i) for i in range(len(crabs_pos))])
```

```python
x = len(crabs_pos) // 2
while True:
    if c(x) > c(x - 1):
        x -= 1
    elif c(x) > c(x + 1):
        x += 1
    else:
        break

c(x)
```
