# Challenge 1: find initial velocity that reaches highest Y position and eventually lies in the target area

```python
input = open("input.txt", "r").read()[:-1]
target_area_a = (207, -115)
target_area_b = (263, -63)
```

```python
target_area_a = (20, -10)
target_area_b = (30, -5)
```

```python
def step(pos, vel):
    x, y = pos
    xv, yv = vel
    x += xv
    y += yv
    xv -= 1 if xv > 0 else (0 if xv == 0 else -1)
    yv -= 1
    return (x, y), (xv, yv)
```

```python
def loss(vel):
    pos = (0, 0)
    max_height = 0
    in_target = False
    while True:
        pos, vel = step(pos, vel)
        max_height = max(max_height, pos[1])
        if not in_target and target_area_a[0] <= pos[0] <= target_area_b[0] and target_area_a[1] <= pos[1] <= target_area_b[1]:
            in_target = True
        if pos[1] < target_area_a[1] and vel[1] <= 0:
            break
    if not in_target:
        return None
    else:
        return max_height
        
loss((6, 9))
```

```python
def optimize():
    best_l = 0
    best_x = 0
    best_y = 0
    for x in range(50):
        for y in range(-100, 200):
            l = loss((x, y))
            if l is None:
                continue
            if l > best_l:
                best_l = l
                best_x = x
                best_y = y
    return best_x, best_y, best_l
```

```python
optimize()
```

# Challenge 2: find every velocity setting that allows the probe to reach the target area

```python
def find_all():
    n = 0
    for x in range(265):
        for y in range(-115, 200):
            l = loss((x, y))
            if l is None:
                continue
            n += 1
            print(x, y)
    return n
```
